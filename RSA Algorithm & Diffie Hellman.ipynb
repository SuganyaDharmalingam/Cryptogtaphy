```python
#Problem 2:

def EEA(a,b):
    x0=1;x1=0;y0=0;y1=1;sign=1;m=0;
    while b!=0:
        r=a%b # remainder
        q=a//b 
        a=b
        b=r
        I=x1
        J=y1
        x1=q*x1+x0
        y1=q*y1+y0
        x0=I
        y0=J
        sign=-sign
        m= m+1
    x=sign*x0
    y=-(sign)*y0
    gcd=a
    return(gcd,x,y,m)
```

```python
[g,x,y,m]= EEA(a,n)
print(EEA(a,n))
print(EEA(b,n))
```
